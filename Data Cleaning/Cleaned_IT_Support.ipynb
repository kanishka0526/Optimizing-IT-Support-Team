```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
)
```

```python
df.columns
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["ticket_id", "ticket_description"])
```

```python
categorical_cols = ["priority", "status", "category", "channel"]

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")
```

```python
df["ticket_description"] = (
    df["ticket_description"]
    .str.lower()
    .str.strip()
    .str.replace(r"[^a-zA-Z0-9 ]", "", regex=True)
)
```
